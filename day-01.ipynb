## Error Handling


Try

Except 

```python
def sum(num1: int , num2:  int) -> int:
    """ 
    A function to add 2 numbes
    """
    
    return num1 + num2

try:
    num1 = int(input("Ente First number: "))
    num2 = int(input("Ente Second number: "))
    
    print(sum(num1 , num2))

except ValueError as err:
    print(f"Got error: {err} \n Enter a integer value and try Again")
except:
    print("Try again and enter the proper value")
```
