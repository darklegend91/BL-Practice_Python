## Find Execution Time of a program

```python
import time
x= 45
y = 56

start_time = time.time()
sum = x+y
print("Sum of x and y is " , sum)
print("Time taken to print is " , (time.time() - start_time) * 10 **6)

start_time = time.time()
sum = x+y
print(f"Sum of {x} and {y} is {sum}")
print("Time taken to print is " , (time.time() - start_time) * 10 **6)
```

```python
# Nested Exception Handling

try:
    num1= int(input("Enter a number: "))
    num2 =  int(input("Enter the divisor: "))

    try:
        result = num1/num2
        print("Result:", result)

    except ZeroDivisionError:
        print("Dividing by 0")

except ValueError:
    print("Please enter a valid integer value.")
```

### Find Generation

```python
try:
    yob= int(input("Enter Your year of birth: "))
    
    if (1946 <= yob <=1964):
        print(f"You are a Baby Boomer as you are born is {yob}.")
    
    elif (1965 <= yob <=1980):
        print(f"You are a Baby Boomer as you are born is {yob}.")
    
    elif (1981 <= yob <=1996):
        print(f"You are a Baby Boomer as you are born is {yob}.")
    
    elif (1997 <= yob <=2012):
        print(f"You are a Baby Boomer as you are born is {yob}.")
    
    elif (2013 <= yob <=2025):
        print(f"You are a Baby Boomer as you are born is {yob}.")
    
    else:
        print(f"Enter the age from 1964 t0 2025 and try Again.")
        
except ValueError as err:
    print(f"Error ocuured : {err} \n Enter an int value ")
except:
    print("Enter proper value and then retry.")
```

### Time Conversion

```python
try:
    time_sec = int(input("Enter number of seconds: "))
    
    hours = int(time_sec // 3600)
    time_rem = time_sec % 3600
    
    minutes = int(time_rem // 60)
    time_rem =  time_rem % 60
    
    print(f"Time is {hours} hours, {minutes} minutes and {time_rem} seconds .")
    
except ValueError as err:
    print(f"Error Ocuured as {err}")
```
