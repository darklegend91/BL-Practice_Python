### Bill Generation

```python
# Step 1: Setting Constants for Fruits Apple GST as 12% and Orange GST as 5%
APPLE_GST = 0.12  # 12% GST
ORANGE_GST = 0.05  # 5% GST

# Step 2: Enter Buyer Name, Price for Apple and Orange per kg, and quantity
# of Apples and Oranges in kg purchased by the buyer
buyer_name = input("Enter Buyer Name: ")
apple_price_kg = int(input("Enter Apple price per kg: "))
apple_quantity_kg = float(input("Enter Apple Quantity in kg: "))
orange_price_kg = int(input("Enter Orange price per kg: "))
orange_quantity_kg = float(input("Enter Orange Quantity in kg: "))

# Step 3: Calculate Total Price of Apple and Orange
total_price_apple = apple_price_kg * apple_quantity_kg
total_price_orange = orange_price_kg * orange_quantity_kg

# Step 4: Calculate the GST Amount for Apple and Orange
total_gst_apple = total_price_apple * APPLE_GST
total_gst_orange = total_price_orange * ORANGE_GST

# Step 5: Calculate the total billing amount for apple and oranges
# inclusive of GST
total_billing_apple = total_price_apple + total_gst_apple
total_billing_orange = total_price_orange + total_gst_orange

# Step 6: Calculate total amount to be paid by the buyer and round it
total_amount = total_billing_apple + total_billing_orange
total_round_amount = round(total_amount)

# Print the bill
print(f"\nBuyer Name: {buyer_name}")
print(f"{'~' * 74}")

print(
    f"| {'Item Code':^10} | {'Price/Unit':^10} | {'# unit':^5}",
    f"| {'Price':^5} | {'GST':^10} | {'Total w/ GST':^10} |"
)

print(f"{'~' * 74}")

print(
    f"| {'Apple':^10} | {'Rs ' + str(apple_price_kg):^10} | {apple_quantity_kg:^5}", # Carrot ^ Symbol set the data in centre in goven width like for appple width is 10
    f"| {'Rs ' + str(total_price_apple):^5} | {'Rs ' + str(total_gst_apple):^10}",
    f"| {'Rs ' + str(total_billing_apple):^10} |"
)

print(
    f"| {'Orange':^10} | {'Rs ' + str(orange_price_kg):^10} | {orange_quantity_kg:^5}",
    f"| {'Rs ' + str(total_price_orange):^5} | {'Rs ' + str(total_gst_orange):^10}",
    f"| {'Rs ' + str(total_billing_orange):^10} |"
)

print(f"{'~' * 74}")
print(f"Total {' ' * 54} \u20B9 {total_amount:.2f}")
print(f"Total Round {' ' * 48} \u20B9 {total_round_amount:.2f}")
print(f"{'~' * 74}")
```

### School Report card
With Custom Exception handling

```python
class InvalidMarksError(Exception):
    """
    Value Entered must be less than or equal to the maximum marks
    """

#Step1 : Define Constatnts
School_Name = "New School Of Learning"
Class_Name = "Class VI"
Total_Subject_Marks = 50
Number_Students = 3

#Step2 : Get Student Name and and Physics and Chemistry Marks for Student 1
try:
    
    student1 = input("Enter name of student 1: ")
    student1_phys_marks = int(input("Enter Physics marks out of 50 for student 1: "))
    if ( student1_phys_marks >50):
        raise InvalidMarksError("Marks cannot be greater than 50.")
    student1_chem_marks = int(input("Enter Chemistry marks out of 50 for student 1: "))
    if ( student1_chem_marks >50):
            raise InvalidMarksError("Marks cannot be greater than 50.")
    student1_math_marks = int(input("Enter Mathematics marks out of 50 for student 1: "))
    if ( student1_math_marks >50):
            raise InvalidMarksError("Marks cannot be greater than 50.")
    total_marks = (student1_phys_marks + student1_chem_marks + student1_math_marks)
    
except InvalidMarksError as err:
    print(f"Error happened as {err}- Enter marks less than {Total_Subject_Marks}")
    
except ValueError as err:
    print(f"Error happened as {err}- Enter marks in float or int format")

else :
    
#Step3: Calculate the precentage of each subject
    student1_phys_perc = round((student1_phys_marks/total_marks *100), 2)
    student1_chem_perc = round((student1_chem_marks/total_marks *100), 2)
    student1_math_perc = round((student1_math_marks/total_marks *100), 2)

#Step 4 : Print For Student 1

    print(f"\n{School_Name} - {Class_Name}- {student1}")
    print("-" * 74)

    print(
        f"| {'Subject':^15}" 
        f"|{'Total Marks':^15}"
        f"|{'Marks Obtained':^15}"
        f"|{'Percentage':^15}"
        )

    print('-' * 74)

    print(
        f"| {'Physics':^15} "
        f"| {Total_Subject_Marks:^15} "
        f"| {student1_phys_marks:^15} "
        f"| {student1_phys_perc:^15} |"
    )

    print(
        f"| {'Chemistry':^15} "
        f"| {Total_Subject_Marks:^15} "
        f"| {student1_chem_marks:^15} "
        f"| {student1_chem_perc:^15} |"
    )

    print(
        f"| {'Mathematics':^15} "
        f"| {Total_Subject_Marks:^15} "
        f"| {student1_math_marks:^15} "
        f"| {student1_math_perc:^15} |"
    )

    print(f"{'-' * 72}")

    overall_percentage = round(
        (total_marks / (Total_Subject_Marks * 3)) * 100, 2
    )

    print(f"Total Marks Obtained: {total_marks}/{Total_Subject_Marks}")
    print(f"Overall Percentage: {overall_percentage:.2f}%")
```
